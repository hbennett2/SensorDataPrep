## TASK ONE

```python
"""Reads sensor data and temperature values into a list"""
def read_files(sensor, temperature):
    
    with open(sensor, 'r') as file:                                  # open the file
        lines = file.readlines()                                     # read the file into list
    file.close()                                                     # close file
    sensor_list = [float(line.strip()) for line in lines]  # convert the strings to floats
    
    with open(temperature, 'r') as file2:
        lines2 = file2.readlines()
    file2.close()
    temp_list = [round(float(line.strip()), 2) for line in lines2]

                                                                     # returns lists
    return sensor_list, temp_list

sensor_list, temp_list = read_files('sensor.txt', 'temperature.txt')
print(sensor_list, "\n")
print(temp_list)
```

## TASK THREE

```python
""" Function creates frequencies from list and converts to dict"""
def get_freq(my_list):
    freq_dict = {}
    for value in my_list:
        if value in freq_dict:
            freq_dict[value] += 1
        else:
            freq_dict[value] = 1
    return freq_dict

# prints frequencies of lists
sensor_freq = get_freq(sensor_list)
print("Sensor Frequencies\n", sensor_freq, "\n")

temp_freq = get_freq(temp_list)
print("Temp Frequencies\n", temp_freq)
```

## TASK SIX

```python
""" Function cleans data """
def clean_list(mylist):
    clean = []
    relate = []  # used to filter out modified values for corresponding list
    for value in mylist:
        if value == 0.0:              # rids of 0.0
            relate.append(True)
            continue
        elif value < 0:
            clean.append(abs(value))  # makes -# an absolute value
        elif value < 1000:            # makes anything from 1-999 become 1000
            clean.append(1000)
        elif value > 2000:            # makes anything 2001+ become 2000
            clean.append(2000)
        else:
            clean.append(value)       # adds valid values to lists
        relate.append(False)
    return clean, relate
```

```python
""" TESTING FUNCTION """
test = [-1000, 0.0, 1000, 1500, 2000, 3000, 0.0]
test1 = [1, 2, 3, 4, 5, 6, 7]
testclean, relate = clean_list(test)
testclean1 = [value for value, drop in zip(test1, relate) if not drop]

print('Clean Test',testclean)
print('Clean Test1',testclean1, '\n')

cleansensor_list, relate = clean_list(sensor_list)
cleantemp_list = [value for value, drop in zip(temp_list, relate) if not drop]

print('Sensor', sensor_list[:5])
print('Clean Sensor',cleansensor_list[:5])
slength= len(cleansensor_list)                # get and print length
print('Length:', slength, 'values\n')
print('Temp', temp_list[:5])
print('Clean Temp', cleantemp_list[:5])
tlength= len(cleantemp_list)
print('Length:', tlength, 'values\n')
```
